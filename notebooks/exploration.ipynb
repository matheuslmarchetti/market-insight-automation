## Como obter sua API key do FRED:

1. **Acesse:** https://fred.stlouisfed.org
2. **Clique em "My Account"** (canto superior direito)
3. **Cadastre-se gratuitamente** (se ainda não tiver conta)
4. **Vá para "Account Settings" → "API Keys"**
5. **Clique em "Request API Key"**
6. **Preencha o formulário** (uso pessoal/educacional é gratuito)

## Depois de obter sua chave:

```python
from fredapi import Fred

# Substitua com SUA chave
fred = Fred(api_key="sua_chave_aqui_32_caracteres")

# Buscar dados de inflação
inflation = fred.get_series("CPIAUCSL")
print(inflation)
```

## Códigos úteis do FRED:
- `CPIAUCSL`: Índice de Preços ao Consumidor (inflação)
- `UNRATE`: Taxa de desemprego
- `GDP`: PIB
- `FEDFUNDS`: Taxa de juros dos Fed Funds

A API key gratuita tem limite de 1.200 requisições por hora, o que é suficiente para a maioria dos usos pessoais e educacionais.

````python
from fredapi import Fred

with open(r"..\..\API_key_FRED.txt", 'r') as f:
    api_key = f.read().strip()

fred = Fred(api_key=api_key)
````

````python
inflation = fred.get_series("CPIAUCSL")

display(inflation.shape)

display(inflation.head(10))

display(inflation.tail(10))

display(inflation.info())

display(inflation.describe())
````
